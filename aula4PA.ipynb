```python
# variaveis que recebe o valor das notas

# Entrada
nota1 = 7
nota2 = 10
nota3 = 8
nota4 = 5
```

```python
# Processamento
media_aritmedica = (nota1 + nota2 + nota3 + nota4) / 4
```
