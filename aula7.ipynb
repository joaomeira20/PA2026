```python
#entrada
nome = str(input("Digite seu nome: "))

b1 = float(input("digite a primeira nota do aluno: "))
while b1 < 0 or b1 > 10:
  print ("nota invalida, digite uma nota de 0 a 10:")
  b1 = float(input("digite a primeira nota do aluno: "))

b2 = float(input("digite a segunda nota do aluno: "))
while b2 < 0 or b2 > 10:
  print ("nota invalida, digite uma nota de 0 a 10:")
  b2 = float(input("digite a segunda nota do aluno: "))

b3 = float(input("digite a terceira nota do aluno: "))
while b3 < 0 or b3 > 10:
  print ("nota invalida, digite uma nota de 0 a 10:")
  b3 = float(input("digite a terceira nota do aluno: "))

b4 = float(input("digite a quarta nota do aluno: "))
while b4 < 0 or b4 > 10:
  print ("nota invalida, digite uma nota de 0 a 10:")
  b4 = float(input("digite a quarta nota do aluno: "))


#processamento
nota = (b1 + b2 + b3 + b4) / 4

#saida
if nota < 5:
  print (f"o aluno {nome} foi reprovado")

elif nota <= 5:
  print (f"o aluno {nome} esta de recuperação")

else :
  print (f"o aluno {nome} foi aprovado")
```

```python
#Entrada
compra = float(input(f"digite o valor total da compra: "))
print (f"valor da compra: R$: {compra}")

#Processamento e Saida
if (compra >= 101) and (compra <= 500):
    des = compra * 0.10
    total = compra - des
    print(f"O valor total da sua compra com 10% de desconto é de: {total}")

elif compra > 500:
    des2 = compra * 0.20
    total = compra - des2
    print(f"O valor total da sua compra com 20% de desconto é de: {total}")

else:
    print(f"O valor total da sua compra é de: {compra}")

```
